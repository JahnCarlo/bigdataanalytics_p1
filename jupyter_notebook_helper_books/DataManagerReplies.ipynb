```python
filepath = "outputFindRepliesCaptureQuantity"

contents = []

with open(filepath) as fp:
    entry = []
    line = fp.readline().split()
    for i in line:
        entry.append(i)
    contents.append(entry)
    while line:
        entry = []
        line = fp.readline().split()
        for i in line:
            entry.append(i)
        contents.append(entry)

modified = []

#print(len(contents))

for entry in contents:
    try:
        if(entry[1] > '8' ):
            modified.append(entry)
    except:
        print ("EOF Reached")

#print(len(modified))

names = []
for entry in modified:
    try:
        names.append(entry[0])
    except:
        print ("EOF Reached")
        
#Everything above works        
contents = []

namesA = "outputFindRepliesD"
with open(namesA) as na:
    entry = []
    line = na.readline().split()
    for i in line:
        entry.append(i)
    contents.append(entry)
    while line:
        entry = []
        line = na.readline().split()
        for i in line:
            entry.append(i)
        contents.append(entry)
        
result = []
#Above works
for element in contents:
    try:
        if element[0] in names:
            result.append(element)
    except:
        print("Index out of bounds")


#Clean self replies
cleanedselfreplies = []
for element in result:
    try:
        if(element[0] != element[1]):
            cleanedselfreplies.append(element[0] + ", " + element[1])
    except:
        print("Index out of bounds")
    
#Clean Repeated Values
clean = []
seen = set()
for element in cleanedselfreplies:
    try:
        if element not in seen:
            clean.append(element)
            seen.add(element)
    except:
        print("Index out of bounds")

#Print to screen to use in text file
for element in clean:
    print(element)
```
