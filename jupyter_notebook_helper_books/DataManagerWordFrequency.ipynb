```python
filepath = "wordFreq.txt"
contents = []

with open(filepath) as fp:
    entry = []
    line = fp.readline().split()
    for i in line:
        entry.append(i)
    contents.append(entry)
    while line:
        entry = []
        line = fp.readline().split()
        for i in line:
            entry.append(i)
        contents.append(entry)
        
modified = []

#print(len(contents))

for entry in contents:
    try:
        if(entry[1] > '10' and len(entry) == 2 and "@" not in entry[0] and "+" not in entry[0]):
            modified.append(entry)
    except:
        print ("EOF Reached")
        

sortedList = sorted(modified , key=lambda x: int(x[1]))
goodSorted = list(reversed(sortedList))

#Clean and format into one, get first 1000 to analyze
cleanSorted = []
for element in goodSorted[50:551]:
    try:
        cleanSorted.append(element[0] + ", " + element[1])
    except:
        print("Index out of bounds")


#Print to screen for use
for element in cleanSorted:
    print(element)
```
